Legal Loanwords in Malayalam
Corpus Based Exploratory Analysis

This notebook analyses loanwords in Malayalam legal documents using a manually annotated dataset. The focus is on source language, grammatical function, integration level, and semantic domain.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel("/content/malayalam_legal_loanwords_dataset (2).xlsx")
print(df.shape)
df.head()

```

```python
loanwords = df[df["is_loanword"] == True]
print(loanwords.shape)
```

This visualization shows the distribution of source languages for loanwords in Malayalam legal documents.

```python
loanwords["origin_language"].value_counts().plot(kind="bar")
plt.title("Loanwords by Source Language")
plt.xlabel("Source language")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

This visualization shows the grammatical functions performed by loanwords in Malayalam legal language.

```python
loanwords["grammatical_function"].value_counts().plot(kind="bar")
plt.title("Loanwords by Grammatical Function")
plt.xlabel("Grammatical function")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

This visualization shows the degree of integration of loanwords into Malayalam legal usage.

```python
loanwords["integration_level"].value_counts().sort_index().plot(kind="bar")
plt.title("Loanwords by Integration Level")
plt.xlabel("Integration level")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

This visualization shows the distribution of loanwords across semantic domains in Malayalam legal documents.

```python
loanwords["semantic_domain"].value_counts().plot(kind="bar")
plt.title("Loanwords by Semantic Domain")
plt.xlabel("Semantic domain")
plt.xticks(rotation=45)
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```
